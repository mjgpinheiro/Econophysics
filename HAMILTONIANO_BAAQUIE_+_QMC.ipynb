```python
# âœ… CÃ“DIGO COMPLETO + EXECUÃ‡ÃƒO
import numpy as np
import matplotlib.pyplot as plt
from py_vollib.black_scholes import black_scholes

class BSPathIntegral:
    def __init__(self, sigma=0.022, r=0.04, T=1/365.25):
        self.sigma, self.r, self.T = sigma, r, T

    def sample_paths(self, n_paths=10000, x0=np.log(681.60)):
        paths = []
        dt = self.T / 50
        for _ in range(n_paths):
            path = [x0]
            x = x0
            for _ in range(49):
                dx = (self.r - 0.5*self.sigma**2)*dt + self.sigma*np.sqrt(dt)*np.random.randn()
                x += dx
                path.append(x)
            paths.append(path)
        return np.array(paths)

def quantum_mc_price(paths, K=695.23, r=0.04, T=1/365.25):
    payoffs = np.maximum(np.exp(paths[:,-1]) - K, 0)
    qmc_price = np.mean(payoffs) * np.exp(-r*T)
    return qmc_price

# ðŸŽ¯ EXECUTA AQUI (copiar tudo)
print("ðŸ”¬ FASE 4: HAMILTONIANO BAAQUIE + QMC")
print("="*60)

pi = BSPathIntegral(sigma=0.022, r=0.04)
paths = pi.sample_paths(n_paths=5000)  # rÃ¡pido

# QMC (Path Integral)
qmc_price = quantum_mc_price(paths, K=695.23)

# Black-Scholes baseline
bs_price = black_scholes('c', 681.60, 695.23, 1/365.25, 0.04, 0.022)

mispricing = (qmc_price - bs_price) / bs_price * 100

print(f"S0:  $681.60 | K:  $695.23 (+2%)")
print(f"Ïƒ24h: {0.022*100:.1f}% | r: {0.04*100:.1f}%")
print(f"\nðŸ“Š PREÃ‡OS:")
print(f"BS:     ${bs_price:.4f}")
print(f"QMC:    ${qmc_price:.4f}")
print(f"Mispricing: {mispricing:.2f}%")
print(f"P(ITM):  {(np.mean(np.exp(paths[:,-1])>695.23)*100):.1f}%")

if abs(mispricing) > 5:
    signal = "ðŸŸ¢ BUY CALL" if mispricing > 0 else "ðŸ”´ SELL CALL"
else:
    signal = "âšª NEUTRO"

print(f"\nðŸŽ¯ SIGNAL: {signal}")
print("ðŸ“š Baaquie: Path Integral â†’ H_BS pricing")
print("âœ… FASE 4 IBM READY!")
```
